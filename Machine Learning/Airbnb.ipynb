```python
import pandas as pd
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv('dataset/' + dataset_csv)

# Clean data
del df['Listing URL']
del df['Host URL']
del df['index']
del df['review_date']
del df['Review ID']
del df['Country Code']
del df['Reviewer ID']
del df['Reviewer Name']
del df['Comments']
del df['Listing ID']
del df['Listing Name']
del df['Host ID']
del df['Host Name']
del df['Postal Code']

df.dropna(axis = 0, how = 'any', subset = None, inplace = True)

# Convert non-numeric values
df = pd.get_dummies(df, columns = ['City', 'Country'])

# Print number of rows and cols
print("dataset (rows, cols): ")
print(df.shape)
print("--------------------------------")

# Display the first few rows of the dataframe
# df.head(n = 5)

# Show contents of elemnt 99
print(df.iloc[99])
print("--------------------------------")

# Display column titles
print(df.columns)
print("--------------------------------")

# Statistical summary
print(df.describe())
print("--------------------------------")
```

```python
from sklearn.model_selection import train_test_split

X = df[['Accomodates', 'Bedrooms', 'Beds', 'Square Feet', 'Overall Rating', 'Longitude', 'Latitude', 'Cleanliness Rating', 'Location Rating']]
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10, shuffle = True)

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5)

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
print(y_train.shape)
print(y_test.shape)
print(y_val.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print("model.intercept_ result: ")
print(model.intercept_)
print("--------------------------------")

print("model.coef_ result: ")
print(model.coef_)
print("--------------------------------")

print("model.score result: ")
print(model.score(X_train, y_train))
print("--------------------------------")

model_results = pd.DataFrame(model.coef_, X_train.columns, columns = ['coef'])
print(model_results)
```
